# Loading the libraries

```r
library(tidyverse)
library(dplyr)
library(tidymodels)
```

# reading and wrangling the data set

```r
set.seed(5)
car <- read_csv("cardata.csv") 
colnames(car) <- c("buying", "maint", "doors", "persons", "lug_boot", "safety", "eval")
cols <- c("buying", "maint", "doors", "persons", "lug_boot", "safety", "eval")
car <- mutate(car, buying = as_factor(buying)) %>%
        mutate( maint = as_factor(maint)) %>%
        mutate(doors = as.factor(doors)) %>%
        mutate(persons = as_factor(persons)) %>%
        mutate(lug_boot = as_factor(lug_boot)) %>%
        mutate(safety = as_factor(safety)) %>%
        mutate(eval = as_factor(eval)) 
car
```

# splitting the data into training data and testing data

```r
car_split <-  initial_split(car, prop = 0.75, strata = eval)
car_train <- training(car_split)
car_test <- testing(car_split)

car_train_prop <- car_train %>%
                      group_by(eval) %>%
                      summarize(n = n()) %>%
                      mutate(percent = 100*n/nrow(car_train))

car_prop <- car %>%
                      group_by(eval) %>%
                      summarize(n = n()) %>%
                      mutate(percent = 100*n/nrow(car))

car_train_prop
car_prop

# car_train is the training data set
# car_test is the testing data set
```

# visualise the exploratory data

```r

car_plot_1 <- car_train %>% 
                group_by(buying, eval) %>%
                summarise (count = n()) %>%
                ggplot(aes(x = buying, y = count, fill = eval)) +
                geom_bar(stat = "identity")

                
car_plot_1
```
